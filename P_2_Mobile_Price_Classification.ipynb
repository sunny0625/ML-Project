```python
# Importing Dataset
price_data = pd.read_csv("https://www.dropbox.com/s/iy0w25eunwx2qlq/MobilePriceClassification.csv?dl=1")
```

```python
# Define X
X = price_data[['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g',
       'touch_screen', 'wifi']]
```

```python
# Define y
y = price_data['price_range']
```

```python
# Importing Train Test Model
from sklearn.model_selection import train_test_split
```

```python
# Spliting X & y into Training & Testing Data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 25)
```

```python
X_train
```

```python
# Importing Model to train ML Model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
# Train the ML Model
knn.fit(X_train, y_train)
```

```python
# Testing the Model
y_pred = knn.predict(X_test)
```

```python
# model accuracy
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
