# Train Data

```python
# Importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Define X
X = df[['Age', 'Gender', 'Tiredness', 'Compulsive behavior', 'Panic attacks',
       'Mood swings', 'Obsessive thinking', 'Depression', 'Anxiety',
       'Lack of concentration', 'Rent', 'I live with my parents',
       'I have been hospitalized before for my mental illness']]
```

```python
# Define y
y = df['MentallyIll']
```

```python
# Spliting X & y into Training & Testing Data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 22)
```

```python
X_train
```

```python
X_test.shape
```

```python
# Importing Model to train ML Model
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
# Train the ML Model
lr.fit(X_train.values, y_train)
```
