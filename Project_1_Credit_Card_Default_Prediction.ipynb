```python
# Define y
y = Data_set['Class']
```

```python
# Define X
X = Data_set[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
```

```python
y.shape, X.shape
```

```python
# Importing train test model
from sklearn.model_selection import train_test_split
```

```python
# Spliting X & y into Training & Testing Data
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 12)
```

```python
X_train
```

```python
# Importing DecisionTree Model
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
```

```python
# To train our ML model
dt.fit(X_train, y_train)
```

```python
# Predict the model
y_pred = dt.predict(X_test)
```

```python
# Model Accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, accuracy_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```
